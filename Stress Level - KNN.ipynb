> ## Algorithme KNN 

```python
Xs = KNeighborsClassifier(n_neighbors=5, weights='uniform')
#On considere 5 voisins pour avoir un resultat plus cohrant avec la taille de la BdD

Xs.fit(x_train, np.ravel(y_train,order='C'))
```

```python
#Prediction
pred1 = Xs.predict(x_test)
print (pred1)
```

```python
print (Xs.score(x_test,y_test))
```

```python
k_range = range(1, len(x_test))
```

```python
score = {}
```

```python
for k in k_range:
    Xs = KNeighborsClassifier(n_neighbors=k, weights='uniform')
    
    Xs.fit(x_test, np.ravel(y_test,order='C'))
    score[k]=Xs.score(x_test, y_test)
    
score
```
